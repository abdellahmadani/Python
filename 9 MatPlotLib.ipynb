## Introduction
<p>
<strong>Matplotlib</strong> est une bibliothèque python qui dessine des graphiques. Il s’agit sûrement l’une des bibliothèques python les plus utilisées pour représenter des graphiques en 2D. Elle permet de produire une grande variété de graphiques et ils sont de grande qualité.
</p>
<p>
Le module <b>pyplot</b> de matplotlib est l’un de ses principaux modules. Il regroupe un grand nombre de fonctions qui servent à créer des graphiques et les personnaliser (travailler sur les axes, le type de graphique, sa forme et même rajouter du texte).
</p>

## Courbes
### Tracer une courbe
<p>
Le module Matplotlib doit, d'abord, être chargé pour pouvoir de tracer les courbes :
</p>

```python
import matplotlib.pyplot as plt
plt.grid(True)
plt.xlabel("Villes")
plt.ylabel("Populations")
plt.title("Recensement en 2014")
villes =['Casa', 'Marrakech', 'Fes', 'Tanger', 'Salé', 'Meknes', 'Rabat']
populations=[3359818, 1329850, 1122403, 980553, 928850, 632079,577827]
plt.plot(villes, populations,'r--s')#c:cyan, -:continu, ^:triangle
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Etudiants")
plt.ylabel("Moyennes")
plt.title("Moyenne des étudiants")
plt.grid(True)
et=range(1, 6)
notes =[12.5, 16.75, 10.25, 11.50, 14]
plt.plot(et,notes,"b--s")#b: blue, --:discontinu, s:square
plt.show()
```

<p>
D’une manière générale les fonctions plt.plot attendent des vecteur/matrice, bref des tableaux
de points du plan.
</p>

```python
import matplotlib.pyplot as plt
days = list(range(1,9))
plt.grid(True)
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values)
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)#génère 100 valeurs uniformement réparties entre -5 et 5. Voir le module Numpy
plt.plot(x,np.sin(x)) # on utilise la fonction sinus de Numpy
plt.ylabel('fonction sinus')
plt.xlabel("l’axe des abcisses")
plt.show()
```

### Tracer plusieurs courbes
<p>
Il est possible de superposer les courbes:
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),"r") # on utilise la fonction sinus de Numpy
plt.plot(x, np.cos(x),"g")
plt.ylabel('fonction sinus & cosinus')
plt.xlabel("l’axe des abcisses")
plt.grid(True)
plt.show()
```

Vous pouvez utiliser aussi :

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x), x, np.cos(x))
plt.ylabel('fonction sinus & cosinus')
plt.xlabel("l’axe des abcisses")
plt.show()
```

### Changer le visuel des courbes
<p>
Il est possible de changer la couleur (b g r c m y k w) et la forme (- -- * -. s ^ ) des courbes:
</p>


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'g^')
plt.ylabel('fonction sinus')
plt.xlabel("l’axe des abcisses")
plt.show()
```

### Largeur des courbes
<p>
Vous pouvez changer la largeur des courbes comme ceci:
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-2*np.pi,2*np.pi,100)
plt.plot(x,np.sin(x), linewidth=5) # on utilise la fonction sinus de Numpy
plt.plot(x, np.cos(x), linewidth=30)
plt.ylabel('fonction sinus & cosinus')
plt.xlabel("l’axe des abcisses")
plt.show()
```

### Afficher la grille
<p>
Vous pouvez également ajouter une grille (et des marqueurs :  o + . x * ^):
</p>

```python
import matplotlib.pyplot as plt
plt.grid(True)
plt.xlabel("abscisses")
plt.ylabel("ordonnés")
plt.plot([50,100,150,200], [2,3,7,10], "b--", linewidth=0.8, marker="*")
plt.plot([50,100,150,200], [2,7,9,10], "g--", linewidth=0.8, marker="+")
plt.show()
```

### Ajouter une légende pour les courbes

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
plt.grid(True)

plt.plot(x,np.sin(x),'r', label="Sinus") # on utilise la fonction sinus de Numpy
plt.plot(x, np.cos(x),'g', label="Cosinus")
plt.ylabel('fonction sinus & cosinus')
plt.xlabel("l’axe des abcisses")
plt.legend()
plt.show()
```

### Sauvegarder une courbe dans un fichier
<p>
La méthode <b>savefig</b> qui permet de sauvegarder le graphique dans un fichier. Elle prend tout simplement en paramètre le chemin (relatif ou absolu) où il faut enregistrer le fichier. Plusieurs formats sont supportés, notamment le PDF et le PNG.
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)

plt.plot(x,np.sin(x),'r*') # on utilise la fonction sinus de Numpy
plt.ylabel('fonction sinus')
plt.xlabel("l’axe des abcisses")

plt.savefig("fig100.jpg")
plt.show()

plt.close()
```

## La fonction scatter

<p>
La fonction scatter permet d'afficher un ensemble de points. On donne les abscisses en premier argument et les ordonnées en second argument.
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
X = np.arange(10)
print(X)
Y = X**2
print (Y)
plt.scatter(X, Y)
plt.xlabel("abscisses")
plt.ylabel("ordonnés")
plt.show()
```

```python
import matplotlib.pyplot as plt
abscisses = range(-5,6) #génère des valeurs entre -5 et 5 (6-1)
ordonnees = [x**2 for x in abscisses]
plt.xlabel("x")
plt.ylabel("x**2")
plt.scatter(abscisses, ordonnees, color='g')
```

## La fonction bar
<p>
Comme, scatter et plot, la fonction bar donne une représentation graphique d'une liste de points. Chaque point est représenté sous la forme d'un rectangle :
</p>

```python
plt.bar([1, 2, 3, 4, 5], [16, 36, 10, 4, 25])
plt.show()
```

<p>
On peut ajuster la largeur des bâtons à l'aide de l'argument optionnel <strong>width</strong>, la couleur des batons avec <strong>color</strong> :
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(0, 6.28, 20)
plt.bar(X, np.sin(X), color="red", width=.1)
plt.xlabel("x")
plt.ylabel("sin")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(0, 6.28, 20)
plt.hist(np.sin(X))
plt.xlabel("x")
plt.ylabel("sin")
plt.show()
```
